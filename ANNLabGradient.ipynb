```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(x,y):
  return 2*x*y+y**2+6*x+2*y
```

```python
def grad(x,y):
  df_x=2*y+6
  df_y=2*x+2*y+2

  return df_x,df_y
```

```python
def gradient_descent(int_x,int_y,eta,epoc):
  x = int_x
  y = int_y
  x_new = [x]
  y_new = [y]

  for i in range(epoc):
    grad_x, grad_y = grad(x,y)
    x= x-eta*grad_x
    y= y-eta*grad_y
    x_new.append(x)
    y_new.append(y)

  return x_new,y_new

```

```python
int_x=2
int_y=-2
eta=0.1
epoc=10

x,y = gradient_descent(int_x,int_y,eta,epoc)
print(x)
print(y)
plt.title("gradient Decent")
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
def gradient_asent(int_x,int_y,eta,epoc):
  x = int_x
  y = int_y
  x_new = [x]
  y_new = [y]

  for i in range(epoc):
    grad_x, grad_y = grad(x,y)
    x= x+eta*grad_x
    y= y+eta*grad_y
    x_new.append(x)
    y_new.append(y)

  return x_new,y_new
```

```python
int_x=2
int_y=-2
eta=0.1
epoc=10

x,y = gradient_asent(int_x,int_y,eta,epoc)
plt.title("gradient_ascent")
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

# Newton

```python
hessian = np.array([[0,2],[2,2]])

gradient = np.array(grad(2,-2))
```

```python
def gradient_descent_newton(int_x,int_y,hessian,epoc):
  init = np.array([int_x,int_y])
  x_new = [int_x]
  y_new = [int_y]

  for i in range(epoc):
    gradient = np.array(grad(init[0],init[1]))
    init = init - np.dot(hessian,gradient)
    x_new.append(init[0])
    y_new.append(init[1])
  return x_new,y_new
```

```python
hessian = np.array([[0,2],[2,2]])
int_x=2
int_y=-2
eta=0.1
epoc=2

x,y = gradient_descent_newton(int_x,int_y,hessian,epoc)
plt.title("newton")
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```
