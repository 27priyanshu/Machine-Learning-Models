```python
import matplotlib.pyplot as plt
import numpy as np

def act(x):
  return 1/(1+np.exp(-x))

def bpa_graph(eta,aplhalist,epoc):
  # parameter decleration

  for alpha in aplhalist:
    t= 0.5
    w1=np.array([0.1,0.8])
    w2=np.array([0.4,0.6])
    w3=np.array([0.3,0.9])
    x=np.array([0.35,0.9])
    errorMBpa=[]
    for i in range(epoc):
      # Forward Pass
      o1 = act(np.dot(x, w1))
      o2 = act(np.dot(x, w2))
      o3= act(np.dot(np.array([o1,o2]), w3))
      error = t-o3
      errorMBpa.append(error)
      # Backward Pass
      Del3=o3*(1-o3)*error
      Del2=o2*(1-o2)*Del3*w3[1]
      Del1=o1*(1-o1)*Del3*w3[0]
      w1[0]=alpha*w1[0]+eta*Del1*x[0]
      w1[1]=alpha*w1[1]+eta*Del1*x[1]
      w2[0]=alpha*w2[0]+eta*Del2*x[0]
      w2[1]=alpha*w2[1]+eta*Del2*x[1]
      w3[0]=alpha*w3[0]+eta*Del3*o1
      w3[1]=alpha*w3[1]+eta*Del3*o2
    plt.plot(errorMBpa)
    plt.xlabel("epoc")
    plt.ylabel("error")
    plt.legend(aplhalist,loc="lower right")
  plt.title(f"eta:{eta}")
  plt.show()
```

```python
bpa_graph(0.01,[0.9,1,0.1],10)
```

```python
bpa_graph(0.1,[0.9,1,0.1],10)
```

```python
bpa_graph(0.5,[0.9,1,0.1],10)
```
