```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download("brown")
nltk.download("stopwords")
```

```python
br_tw = nltk.corpus.brown.tagged_words(categories="mystery")
```

```python
br_tw
```

```python
len(br_tw)
```

```python
# br_tw = [(word, tag) for (word, tag) in br_tw if word.isalpha() and word not in stopwords.words('english')]
# In the above code I removed stopwords, but for this specific task I realised that even the stopwords are
# getting certain tags, and the objective is to study these tags only, hence I choose to keep them

br_tw = [(word, tag) for (word, tag) in br_tw if word.isalpha()]
# In the above code I am removing the punctuation, but if the user wants to study the tags of punctuation marks as well
# you may comment out this line and re-run.
```

```python
br_tw
```

```python
len(br_tw)
```

```python
br_ts = nltk.corpus.brown.tagged_sents(categories="mystery")
```

```python
br_ts
```

```python
len(br_ts)
```

Q1. How many words and sentences specific to given tag are there?

```python
tags_words = [b for (a,b) in br_tw]
fd_words = nltk.FreqDist(tags_words)
print(f"tags for words: {fd_words.most_common()}")
```

```python
tags_sents = [b for i in br_ts for (a,b) in i]
fd_sents = nltk.FreqDist(tags_sents)
print(f"tags for sents: {fd_sents.most_common()}")
```

Q2. Compute top 10 POS tags and their counts

```python
pos_tags_10 = [(x,y) for x,y in fd_words.most_common(10)]
pos_tags_10
```

Q3. Compute top 10 words and their counts

```python
top_words = [x for x,y in br_tw]
fd_top_words = nltk.FreqDist(top_words)
fd_top_words.most_common(10)
```

```python
# we get results like 'the', 'to' etc, maybe removing stopwords would give us some words which are
# more meaningful to the sentence.

br_tw_edited = [(word, tag) for (word, tag) in br_tw if word.isalpha() and word not in stopwords.words('english')]
top_words2 = [x for x,y in br_tw_edited]
fd_top_words2 = nltk.FreqDist(top_words2)
fd_top_words2.most_common(25)
```

Q4. How many different POS tags are represented in this brown category.

```python
pos_tags = [x for x,y in fd_words.most_common()]
print(f"Number of different POS tags: {len(pos_tags)}")

pos_tags
```

Q5. What is the most frequent adverb or verb

```python
adverbs = [x for x,y in br_tw if y in ["RB", "RBS", "RBR"]]
adverbs
```

```python
common_adverb = nltk.FreqDist(adverbs)
common_adverb.most_common(10)
```

```python
verbs = [x for x,y in br_tw if y in ["VB", "VBD", "VBG", "VBN", "VBP", "VBZ"]]
verbs
```

```python
common_verb = nltk.FreqDist(verbs)
common_verb.most_common(10)
```

Q6. Consider any word in the corpus, how many tokens are there for that
