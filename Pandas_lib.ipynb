```python
import numpy as np

data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50],
        'C': [100, 200, 300, 400, 500],
        'D': ['a', 'b', 'c', 'd', 'e']}

arr = np.array(list(data.values()))

print(arr)
```

```python
import pandas as pd
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.bar(df['D'], df['C'], color='skyblue')
plt.xlabel('D')
plt.ylabel('C')
plt.title('Bar Plot of C vs D')
plt.show()
```

```python
import seaborn as sns

# Create a scatter plot using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='A', y='B', data=df, hue='D', palette='viridis', s=100)
plt.xlabel('A')
plt.ylabel('B')
plt.title('Scatter Plot of A vs B colored by D')
plt.show()
```

```python
df = pd.DataFrame(data)

# Calculate the total sum of values in column C
total_sum = df['C'].sum()

# Calculate the percentage of each value in column C
df['C_percentage'] = df['C'] / total_sum * 100

# Create a pie chart using Matplotlib
plt.figure(figsize=(8, 6))
plt.pie(df['C_percentage'], labels=df['D'], autopct='%1.1f%%')
plt.title('Pie Chart of C vs D')
plt.show()
```

```python

df.drop('D', axis=1, inplace=True)

# Print the updated DataFrame
print(df)
```
