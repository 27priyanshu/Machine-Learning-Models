## Bigram and Trigram

```python
bigram = {}
trigram = {}
for i in range(len(corpus_lower)-2):
    bigram_pair = (corpus_lower[i], corpus_lower[i+1])
    trigram_pair = (corpus_lower[i], corpus_lower[i+1],corpus_lower[i+2])

    print(bigram_pair)
    if bigram_pair in bigram.keys():
        bigram[bigram_pair] += 1
    else:
        bigram[bigram_pair] = 1
    print(bigram[bigram_pair])

    if trigram_pair in trigram.keys():
        trigram[trigram_pair] += 1
    else:
        trigram[trigram_pair] = 1
    print(trigram[trigram_pair])
```

```python
print(f'Count for Bigram ("the","king") : {bigram[("the","king")]}')
print(f'Count for Trigram ("," , "," , "the") : {trigram[("," , "," , "the")]}')
```

Function by which we can input the text to get next word

```python
def get_next_word(input_text, bigram, trigram, corpus_vocab):
    denominator = bigram[("country","is")]

    count = []
    for word in corpus_vocab:
        if ("country","is",word) in trigram.keys():
            count.append([word, trigram[("country","is",word)]])

    prob_count = []
    for word_count in count:
        prob_count.append([word_count[1] / denominator, word_count[0]])

    prob_count = sorted(prob_count)

    next_word = prob_count[:1]
    next_word = next_word[0] if next_word else None

    return next_word

# Example usage
input_text = "the country is"
result = get_next_word(input_text, bigram, trigram, corpus_vocab)
print(result)
```
