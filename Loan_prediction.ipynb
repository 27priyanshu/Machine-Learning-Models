```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/loan_data.csv")
df
```

## EDA

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['Dependents','Loan_ID','Gender','Education','Self_Employed','Married'],axis=1,inplace=True)
```

```python
df
```

```python
df.fillna(df['Loan_Amount_Term'].median(),axis=1,inplace=True)
df.fillna(df['Credit_History'].median(),axis=1,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Property_Area'] = le.fit_transform(df['Property_Area'])
df['Loan_Status'] = le.fit_transform(df['Loan_Status'])


df
```

```python
y = np.array(df['Loan_Status'])

df.drop(['Loan_Status'],axis=1,inplace=True)

x = df.values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
lr = LinearRegression()

lr.fit(x_train,y_train)

lrp = lr.predict(x_test)
```

```python
lrp=np.clip(np.round(lrp),0,1)
lrp
```

```python
print(100-mean_absolute_error(y_test,lrp)*100)
print(100-mean_squared_error(y_test,lrp)*100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score,confusion_matrix,classification_report,accuracy_score
```

```python
lr = LogisticRegression()

lr.fit(x_train,y_train)

lrp = lr.predict(x_test)
```

```python
lrp
```

```python
print(r2_score(y_test,lrp)*100,'\n')
print(confusion_matrix(y_test,lrp),'\n')
print(classification_report(y_test,lrp),'\n')
print(100*accuracy_score(y_test,lrp),'\n')
```
