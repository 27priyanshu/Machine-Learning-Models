6 Assumptions of linear regression include:

Linearity: The relationship between the dependent and independent variables is linear.<br>

Independence: The observations are independent of each other.<br>

Homoscedasticity: The variance of the errors is constant across all levels of the independent variables.<br>

Normality: The errors follow a normal distribution.<br>

No multicollinearity: The independent variables are not highly correlated with each other.<br>

No endogeneity: There is no relationship between the errors and the independent variables.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
price = fetch_california_housing()
X=price.data
y=price.target
```

```python
X
```

```python
for i in range(len(X[0])):
  plt.scatter(X[:,i],y)
  plt.xlabel(f"{price.feature_names[i]}")
  plt.ylabel("Target")
  plt.show()
```

2nd Assumption - Autocorrelation

```python
df = pd.DataFrame(price.data,columns = price.feature_names)

df["target"]= price.target

corr_matrix = df.corr()

plt.figure(figsize=(8,6))

sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')

plt.show()
```
