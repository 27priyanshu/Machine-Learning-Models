```python
 df = df.replace(to_replace="?", value=np.nan)
```

```python

price = df['price'].loc[df['price']!=np.nan]
pmean = price.astype(str).astype(int).mean()
df['price']=df['price'].replace('?',pmean).astype(int)
df['price'].head()
```
